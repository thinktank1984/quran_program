```python
import os
import pandas as pd
import requests

# File paths
csv_file_path = r"D:\Downloads\quran_download.csv"
download_dir = r"D:\Downloads\quran_files"

# Create directory if it doesn't exist
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

# Read the CSV file
df = pd.read_csv(csv_file_path)

# Function to download a file
def download_file(url, save_path):
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()  # Raise an error for bad status
        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        print(f"Downloaded: {save_path}")
    except Exception as e:
        print(f"Failed to download {url}. Error: {e}")

# Loop through each row in the DataFrame
for index, row in df.iterrows():
    # Download audio
    audio_url = row['audio']
    audio_filename = os.path.join(download_dir, os.path.basename(audio_url))
    download_file(audio_url, audio_filename)

    # Download image
    image_url = row['image']
    image_filename = os.path.join(download_dir, os.path.basename(image_url))
    download_file(image_url, image_filename)
```

```python
import csv
from pydub import AudioSegment
from pydub.silence import detect_nonsilent
import json
import os
from typing import Dict, List, Tuple

print("starting script execution")

class AudioSilenceDetector:
    def __init__(
        self,
        base_dir: str,           # base directory where all files are located
        min_silence_len: int = 500,  # minimum silence length in ms
        silence_thresh: int = -40,    # silence threshold in dB
        padding: int = 100,          # amount of padding to add to non-silent sections
        seek_step: int = 10,          # step size for silence detection in ms
        input_dir: str = "q_files",   # directory containing input audio files
        output_dir: str = "splits"    # directory to save split audio files
    ):
        print(f"\nInitializing AudioSilenceDetector with parameters:")
        print(f"min_silence_len: {min_silence_len}ms")
        print(f"silence_thresh: {silence_thresh}dB")
        print(f"padding: {padding}ms")
        print(f"seek_step: {seek_step}ms")
        
        self.base_dir = base_dir
        self.config = {
            'min_silence_len': min_silence_len,
            'silence_thresh': silence_thresh,
            'padding': padding,
            'seek_step': seek_step,
            'input_dir': input_dir,
            'output_dir': output_dir
        }
        
        # Create absolute paths
        self.input_path = os.path.join(base_dir, input_dir)
        self.output_path = os.path.join(base_dir, output_dir)
        
        # Create output directory if it doesn't exist
        os.makedirs(self.output_path, exist_ok=True)
        
        print(f"Base directory: {self.base_dir}")
        print(f"Input directory: {self.input_path}")
        print(f"Output directory: {self.output_path}")
        print("Initialization complete\n")
        
    def detect_and_split_audio(self, audio_filename: str) -> List[Dict]:
        """
        Detect non-silent segments in an audio file and save them as separate files.
        Returns list of split information including saved file paths.
        """
        print(f"\nProcessing audio file: {audio_filename}")
        try:
            # Construct full path to audio file
            audio_path = os.path.join(self.input_path, audio_filename)
            
            if not os.path.exists(audio_path):
                print(f"Warning: Audio file {audio_path} not found")
                return []
            
            print(f"Loading audio file: {audio_path}")
            try:
                audio = AudioSegment.from_mp3(audio_path)
                print(f"Successfully loaded audio file. Duration: {len(audio)}ms")
            except Exception as e:
                print(f"Error loading audio file {audio_filename}: {str(e)}")
                return []
            
            print("Detecting non-silent ranges...")
            # Detect non-silent ranges
            nonsilent_ranges = detect_nonsilent(
                audio,
                min_silence_len=self.config['min_silence_len'],
                silence_thresh=self.config['silence_thresh'],
                seek_step=self.config['seek_step']
            )
            print(f"Found {len(nonsilent_ranges)} non-silent ranges")
            
            # Add padding to the ranges
            padding = self.config['padding']
            padded_ranges = []
            for start, end in nonsilent_ranges:
                padded_start = max(0, start - padding)
                padded_end = min(len(audio), end + padding)
                padded_ranges.append((padded_start, padded_end))
            print(f"Applied {padding}ms padding to ranges")
            
            # Generate splits info and save audio segments
            splits_info = []
            base_filename = os.path.splitext(audio_filename)[0]
            
            print("Exporting segments...")
            for idx, (start, end) in enumerate(padded_ranges, 1):
                try:
                    print(f"Processing segment {idx}/{len(padded_ranges)}")
                    # Extract the segment
                    segment = audio[start:end]
                    
                    # Generate output filename
                    output_filename = f"{base_filename}_{idx}.mp3"
                    output_path = os.path.join(self.output_path, output_filename)
                    print(f"Exporting to: {output_filename}")
                    
                    # Export the segment
                    segment.export(output_path, format="mp3")
                    
                    # Store split information
                    splits_info.append({
                        'start_ms': start,
                        'end_ms': end,
                        'duration_ms': end - start,
                        'split_file': output_filename
                    })
                    print(f"Successfully exported segment {idx}")
                except Exception as e:
                    print(f"Error exporting segment {idx} of {audio_filename}: {str(e)}")
                    continue
            
            if splits_info:
                print(f"Successfully processed {audio_filename}: created {len(splits_info)} splits")
            return splits_info
            
        except Exception as e:
            print(f"Error processing {audio_filename}: {str(e)}")
            return []

    def process_csv_file(self, csv_filename: str) -> Dict:
        """
        Process all audio files listed in CSV and return a dictionary with split information
        """
        result = {}
        csv_path = os.path.join(self.base_dir, csv_filename)
        
        try:
            print(f"\nReading CSV file: {csv_path}")
            with open(csv_path, 'r') as csvfile:
                reader = csv.reader(csvfile)
                header = next(reader)  # Skip header row
                print(f"CSV header: {header}")
                
                total_rows = sum(1 for _ in reader)  # Count rows
                csvfile.seek(0)  # Reset file pointer
                next(reader)  # Skip header again
                
                print(f"Found {total_rows} rows to process")
                
                for row_num, row in enumerate(reader, 1):
                    print(f"\nProcessing row {row_num}/{total_rows}")
                    if len(row) >= 2:  # Ensure we have both audio and image
                        audio_file, image_file = row[0], row[1]
                        print(f"Audio file: {audio_file}, Image file: {image_file}")
                        splits_info = self.detect_and_split_audio(audio_file)
                        
                        if splits_info:  # Only add to result if processing was successful
                            result[audio_file] = {
                                'image': image_file,
                                'splits': splits_info
                            }
                            print(f"Added results for {audio_file}")
                    else:
                        print(f"Warning: Invalid row in CSV: {row}")
                        
        except Exception as e:
            print(f"Error processing CSV file: {str(e)}")
            
        print(f"\nProcessed {len(result)} files successfully")
        return result

    def save_json_output(self, result: Dict, output_filename: str = 'audio_splits.json'):
        """
        Save the detection results to a JSON file
        """
        output_path = os.path.join(self.base_dir, output_filename)
        print(f"\nSaving results to {output_path}")
        with open(output_path, 'w') as f:
            json.dump({
                'config': self.config,
                'files': result
            }, f, indent=2)
        print("Results saved successfully")

def main():
    # Get the current working directory
    base_dir = r"D:\code\kv-cache"
    print(f"\nWorking directory: {base_dir}")
    
    print("\nInitializing detector...")
    # Initialize detector with optimal values for Quran recitation
    detector = AudioSilenceDetector(
        base_dir=base_dir,
        min_silence_len=700,     # 700ms minimum silence
        silence_thresh=-40,      # -40dB threshold
        padding=100,            # 100ms padding
        seek_step=10,           # 10ms step size
        input_dir="q_files",    # directory containing input audio files
        output_dir="splits"     # directory to save split files
    )
    
    print("\nProcessing CSV file...")
    # Process the CSV file
    result = detector.process_csv_file('quran_download.csv')
    
    print("\nSaving results...")
    # Save results
    detector.save_json_output(result)
    
    print("\nScript execution completed")



# Invoke main() directly for immediate execution
main()
```
