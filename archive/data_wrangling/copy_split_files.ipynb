```python
import os
import shutil

def process_files(csv_path, source_dir, dest_dir):
    # Ensure destination directory exists
    os.makedirs(dest_dir, exist_ok=True)
    
    # Read the file line by line
    with open(csv_path, 'r', encoding='utf-8') as file:
        for line in file:
            # Split each line by comma
            parts = line.strip().split(',')
            if len(parts) != 2:
                print(f"Skipping invalid line: {line}")
                continue
            
            # Get file names and clean whitespace
            file_to_delete = parts[0].strip()
            file_to_copy = parts[1].strip()
            
            # Setup full paths
            delete_path = os.path.join(dest_dir, file_to_delete)  # Changed to dest_dir
            source_path = os.path.join(source_dir, file_to_copy)
            dest_path = os.path.join(dest_dir, file_to_copy)
            
            print(f"\nProcessing row:")
            print(f"Delete from q_files: {delete_path}")
            print(f"Copy from splits: {source_path}")
            print(f"Copy to q_files: {dest_path}")
            
            # Copy the new file
            if os.path.exists(source_path):
                shutil.copy2(source_path, dest_path)
                print(f"Successfully copied: {source_path} to {dest_path}")
            else:
                print(f"ERROR: Source file not found: {source_path}")
            
            # Delete the old file
            if os.path.exists(delete_path):
                os.remove(delete_path)
                print(f"Successfully deleted: {delete_path}")
            else:
                print(f"ERROR: File to delete not found: {delete_path}")

# Define paths
csv_path = r"D:\code\kv-cache\split.csv"
source_dir = r"D:\code\kv-cache\splits"
dest_dir = r"D:\code\kv-cache\q_files"

# Process the files
process_files(csv_path, source_dir, dest_dir)
```
